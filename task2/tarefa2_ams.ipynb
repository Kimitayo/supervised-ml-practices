## 1.2. Organizando os tipos numéricos


```python
# Obtendo os nomes das colunas numéricas
tipos_numericos = ['int64', 'float64']
att_numericos2 = dataset2.select_dtypes(include=tipos_numericos)
att_numericos3 = dataset3.select_dtypes(include=tipos_numericos)

# Selecionando as colunas
att_numericos2 = list(att_numericos2.columns)
att_numericos3 = list(att_numericos3.columns)

rotulo = dataset1['class'] # Pegando o rótulo

dataset_numerico1 = dataset1.iloc[:, :-1] # Obtendo todas as colunas, exceto a última
dataset_numerico2 = dataset2[att_numericos2]
dataset_numerico3 = dataset3[att_numericos3]
```

# **2. Correlação de Pearson**

```python
import seaborn as sn
import matplotlib.pyplot as plt

# Calcular a correlação
pearson1 = dataset_numerico1.corr(method='pearson')
pearson2 = dataset_numerico2.corr(method='pearson')
pearson3 = dataset_numerico3.corr(method='pearson')
```

## 2.1. Dataset 1



### a) Mostrar os top 20 mais correlados com limiar de 0,8

```python
# Desempilhar em formato de pares
corr_pairs1 = pearson1.unstack()

# Remover self-correlations (idade x idade)
corr_pairs1 = corr_pairs1.drop(labels=[(col, col) for col in pearson1.columns])

# Remover duplicados
corr_pairs1 = corr_pairs1.drop_duplicates()

# Ordenar pelo valor absoluto
sorted_corr1 = corr_pairs1.abs().sort_values(ascending=False)
filtra1 = sorted_corr1[sorted_corr1 >= 0.8]

# Pegar os 20 mais fortes
top20_dataset1 = filtra1.head(20)
print(top20_dataset1)
```

### b) Plotar heatmap

```python
import matplotlib.pyplot as plt

# Pegar apenas os atributos do top 20
att_top20_1 = list(set([par[0] for par in top20_dataset1.index] + [par[1] for par in top20_dataset1.index]))

# Filtrar a matriz de correlção com esses att do top 20
corr_top20_1 = pearson1[att_top20_1].loc[att_top20_1]

plt.figure(figsize=(5,5))
plt.rcParams.update({'font.size': 11})
ax = sn.heatmap(corr_top20_1, annot = True, fmt = '.1f', square=True, linewidths= .3)
plt.show()
```

### c) Removendo com limir de 0,8

```python
import numpy as np

# Declarando a limiar
threshold = 0.8

# Evitar pares duplicados
upper1 = pearson1.where(np.triu(np.ones(pearson1.shape), k=1).astype(bool))

# Remove a segunda coluna de cada par com correlação >= 0.8
to_drop1 = [col for col in upper1.columns if (upper1[col] >= threshold).any()]

# Dataframe final
df_final1 = dataset_numerico1.drop(columns=to_drop1)
# print(df_final1)

# Visualizar o que foi removido
for col in upper1.columns:
    pos_pairs = list(upper1.index[upper1[col] >= threshold])
    neg_pairs = list(upper1.index[upper1[col] <= -threshold])
    for c in pos_pairs:
        print(f"Removendo: {col} (ρ={pearson1.loc[c, col]:.3f} com {c})")
    for c in neg_pairs:
        print(f"Mantendo (negativa): {col} (ρ={pearson1.loc[c, col]:.3f} com {c})")

print(f"\nTotal removidas: {len(to_drop1)}")
print(f"Shape original:  {dataset_numerico1.shape}")
print(f"Shape final:     {df_final1.shape}")

# Salvar
df_final1.to_csv('base1.csv', index=False)
```

# **3. PCA**

## 3.1. Aplicando a normalização nos datasets

```python
from sklearn.preprocessing import StandardScaler

# Dataset 1
X_std1 = StandardScaler().fit_transform(dataset_numerico1)
print(X_std1)
print(X_std1.shape)
```

```python
# Dataset 2
X_std2 = StandardScaler().fit_transform(dataset_numerico2)
print(X_std2)
print(X_std2.shape)
```

```python
# Dataset 3
X_std3 = StandardScaler().fit_transform(dataset_numerico3)
print(X_std3)
print(X_std3.shape)
```

## 3.2. Adaptando os datasets para raiz de n

##

```python
# Dataset 1

from sklearn.decomposition import PCA

# raiz de n componentes
n_componentes1 = int(np.sqrt(dataset_numerico1.shape[0]))

# Tinha dado erro, por iso ter que limitar ao máximo permitido
max_components1 = min(X_std1.shape[0], X_std1.shape[1])
n_componentes1 = min(n_componentes1, max_components1)

# Aplicar PCA
pca1 = PCA(n_components=n_componentes1)
pca_result1 = pca1.fit_transform(X_std1)

print(pca_result1)
print(pca_result1.shape)
```

```python
# Dataset 2

# Raiz componentes
n_componentes2 = int(np.sqrt(dataset_numerico2.shape[0]))

# Tinha dado erro, por iso ter que limitar ao máximo permitido
max_components2 = min(X_std2.shape[0], X_std2.shape[1])
n_componentes2 = min(n_componentes2, max_components2)

# Aplicar PCA
pca2 = PCA(n_components=n_componentes2)
pca_result2 = pca2.fit_transform(X_std2)

print(pca_result2)
print(pca_result2.shape)
```

```python
# Dataset 3

# Raiz de n componentes
n_componentes3 = int(np.sqrt(dataset_numerico3.shape[0]))

# Tinha dado erro, por iso ter que limitar ao máximo permitido
max_components3 = min(X_std3.shape[0], X_std3.shape[1])
n_componentes3 = min(n_componentes3, max_components3)

# Aplicar PCA
pca3 = PCA(n_components=n_componentes3)
pca_result3 = pca3.fit_transform(X_std3)

print(pca_result3)
print(pca_result3.shape)
```

## 3.3. Aplicando PCA com 90% de variância

```python
# Dataset 1

# Indicar a porcentagem
pca1_90 = PCA(n_components=0.90, whiten=True)

# Execução do PCA
pca_result1_90 = pca1_90.fit_transform(X_std1)

print(pca_result1_90)
print(pca_result1_90.shape)
```

```python
# Dataset 2

pca2_90 = PCA(n_components=0.90, whiten=True)
pca_result2_90 = pca2_90.fit_transform(X_std2)
print(pca_result2_90)
print(pca_result2_90.shape)
```
